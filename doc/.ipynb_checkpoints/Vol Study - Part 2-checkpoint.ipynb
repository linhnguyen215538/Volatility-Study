### a. Gather data from Yahoo Finance

```python
driver = webdriver.Chrome() 
ticker_list = ['SPY','IWM','DIA','TLT','QQQ','GLD']
r=0.0012      # risk free rate

#Get dividends list
q_dict = []
S0_dict = []

for ticker in ticker_list:
    driver.get('https://finance.yahoo.com/quote/'+ticker+'?p='+ticker)
    
    WebDriverWait(driver, 15).until(EC.visibility_of_element_located((By.XPATH, '/html/body/div[1]/div/div/div[1]/div/div[3]/div[1]/div/div[1]/div/div/div/div[2]/div[2]/table/tbody/tr[6]/td[2]/span')))
    q= driver.find_element_by_xpath('/html/body/div[1]/div/div/div[1]/div/div[3]/div[1]/div/div[1]/div/div/div/div[2]/div[2]/table/tbody/tr[4]/td[2]/span').text
    if q =='N/A (N/A)':
        q = '(0.00%)'
    q = float(q[0:-1])/100 
    q_dict.update({ticker:q})
    
    WebDriverWait(driver, 15).until(EC.visibility_of_element_located((By.XPATH, "//span[@class='Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(ib)']")))
    S0 = driver.find_element_by_xpath("//span[@class='Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(ib)']").text
    S0 = float(S0.replace(',',''))
    S0_dict.update({ticker:S0})
    
driver.close()
```

### b. Create list of expiration dates and Time to maturity T

```python
# Expirations that don't exist in Yahoo: date(2021,2,19), date(2021, 4,16), date(2021, 5, 21), date(2021,7, 16)

today = date.today()
expiration_list = [date(2020,8,21),date(2020,9,18), date(2020,10,16),date(2020,11,20),date(2020,12,18),
                  date(2021,1,15),date(2021,3, 19), date(2021, 6,18)]
expiration_str = ['August 21, 2020','September 18, 2020','October 16, 2020', 'November 20, 2020', 'December 18, 2020',
                 'January 15, 2021','March 19, 2021', 'June 18, 2021']
expiration_urls = ['1597968000','1600387200', '1602806400','1605830400','1608249600',
                  '1610668800','1616112000','1623974400']

Tlist = []
for i in expiration_list:
    Tlist.append((i - today).days/365)
```

# Part 2. Define functions to use

### a. Black Scholes & Implied Vol

```python
def BlackScholes(S,K,r,q, T,sigma, option):

    d1= (np.log(S/K) + (r-q+sigma**2/2)*T)/(sigma*np.sqrt(T))
    d2= d1-sigma*np.sqrt(T)

    value = 0
    if option =="Call":
        return S*np.exp(-q*T)*norm.cdf(d1)-np.exp(-r*T)*K*norm.cdf(d2)
    elif option =="Put": 
        return -S*np.exp(-q*T)*norm.cdf(-d1)+np.exp(-r*T)*K*norm.cdf(-d2)

def impliedVol(Opt_price, S0,K,r,q,T, option, vol_guess):
    try:
        return newton(func = lambda sigma:BlackScholes(S0,K,r,q,T, sigma, option = option)-Opt_price,
                      x0 = vol_guess, tol = 1e-8)
    except:
        try:
            return newton(func = lambda sigma:BlackScholes(S0,K,r,q,T, sigma,option = option)-Opt_price,
                  x0 = vol_guess)
        except:
            return vol_guess
```

### b. Define Vol_calc function: Get options data from Yahoo; Calculate Implied Vol for ATM and OTM strikes using Newton function

```python
dec_atm_yahoo = []
dec_atm_newton = []

def vol_calc(ticker, option):
    S0 = S0_dict.get(ticker)
    q = q_dict.get(ticker)
    
#Get Options data
    for time in expiration_str:         
        driver.get('https://finance.yahoo.com/quote/'+ticker+'/options?p='+ticker)
        WebDriverWait(driver, 15).until(EC.visibility_of_element_located((By.XPATH, '/html/body/div[1]/div/div/div[1]/div/div[3]/div[1]/div/div[2]/div/div/section/section[2]/div[1]/span[1]/span')))
        links = driver.find_element_by_xpath("//select[@class='Fz(s)']")
        
        if time not in links.text:
            continue      #skip time if option expiration is not available in Yahoo
        
        T = Tlist[expiration_str.index(time)]    #set T = time to expiration
        expi_url = expiration_urls[expiration_str.index(time)]
        
        driver.get('https://finance.yahoo.com/quote/'+ticker+'/options?p='+ticker+'&date='+expi_url)
        WebDriverWait(driver, 15).until(EC.visibility_of_element_located((By.XPATH, '/html/body/div[1]/div/div/div[1]/div/div[3]/div[1]/div/div[2]/div/div/section/section[2]/div[1]/span[1]/span')))
     
        #Create a handle, page, to handle the contents of the website
        page = requests.get(driver.current_url)
        #Store the contents of the website under doc
        doc = lh.fromstring(page.content)
        #Parse data that are stored between <tr>..</tr> of HTML
        tr_elements = doc.xpath('//tr')
        [len(Tr) for Tr in tr_elements[:12]]
        tr_elements = doc.xpath('//tr')
        #Create empty list
        col=[]
        i=0
        #For each row, store each first element (header) and an empty list
        for t in tr_elements[0]:
            i+=1
            name=t.text_content()
            #print('%d:"%s"'%(i,name))
            col.append((name,[]))
            
        #Since out first row is the header, data is stored on the second row onwards
        for j in range(1,len(tr_elements)):
            #T is our j'th row
            Tr=tr_elements[j]

            #If row is not of size 11, the //tr data is not from our table 
            if len(Tr)!=11:
                break
            #i is the index of our column
            i=0
            #Iterate through each element of the row
            for t in Tr.iterchildren():
                data=t.text_content() 
                #Check if row is empty
                if i>0:
                #Convert any numerical value to integers
                    try:
                        data=int(data)
                    except:
                        pass
                #Append the data to the empty list of the i'th column
                col[i][1].append(data)
                #Increment i for the next column
                i+=1

#Process DF and print out to csv
        Dict={title:column for (title,column) in col}
        df=pd.DataFrame(Dict) 
        
        df = df.drop(index = df[df['Bid']=='-'].index)
        df = df.loc[df[df['Contract Name']=='Contract Name'].index.tolist()[0]+1:]
        df['Strike'] = [float(x.replace(',','')) for x in df['Strike']]
        df['Bid'] = [float(x.replace(',','')) for x in df['Bid']]
        df['Implied Volatility'] = [float(x[:-1]) for x in df['Implied Volatility']]
        df['Implied Volatility'] = df['Implied Volatility']/100
        df = df.reset_index(drop = True)
    
#Set up bounds for Strikes Out of the Money (S0Â± 10%)
        K_atm= int(S0)  
        if option == 'Call':
            K_otm = int(S0 *1.1)
        else: 
            K_otm = int(S0*0.9)
            
        Klist = []
        Klist.append(K_atm), Klist.append(K_otm)
    
        siglist = [0]*2
        siglist[0]=df.loc[df['Strike']==K_atm]['Implied Volatility'].values
        siglist[1]=df.loc[df['Strike']==K_otm]['Implied Volatility'].values
        
        Clist = [0]*2
        Clist[0]=df.loc[df['Strike']==K_atm]['Bid'].values
        Clist[1]=df.loc[df['Strike']==K_otm]['Bid'].values
        
        BS_put = [BlackScholes(S0,K,r,q,T, sigma, option = option) for K,sigma in zip(Klist, siglist)]
        Diff = df[df['Strike']==K_atm]['Bid'] - BS_put[0]

#Calculate Implied vol from Newton
        Newton = []                       
        for v in [impliedVol(C, S0,K,r,q,T, option = option, vol_guess =sig) for C,K,sig in zip(Clist, Klist, siglist)]:
            Newton.append(v)
        
        output = pd.DataFrame({'Yahoo IV': [siglist[0][0] if len(siglist[0])!=0 else [], siglist[1][0] if len(siglist[1])!=0 else []], 
                'Newton IV': [Newton[0][0] if len(Newton[0])!= 0 else [], Newton[1][0] if len(Newton[1])!=0 else []]}, 
                index=['ATM', 'OTM'])
        print(ticker + ' expiring on '+ time)
        print(output)
        
        if time =='December 18, 2020':
            dec_atm_yahoo.append(siglist[0][0] if len(siglist[0])!=0 else 0)
            dec_atm_newton.append(Newton[0][0] if len(Newton[0])!= 0 else 0)
```

# Part 3. Call vol_calc function for all tickers in list and print out ATM & OTM Implied Vol calculated from Newton function

```python
for ticker in ticker_list:
    driver = webdriver.Chrome()
    vol_calc(ticker,option = 'Put')
    driver.close()

print("Dec ATM Yahoo vol: ", dec_atm_yahoo)
print("Dec ATM Newton vol: ", dec_atm_newton)
```

# Part 4. Calculating 1 month realised vol for different time periods

Time periods concerned:

2011-01-01 - 2018-01-21: recovering from 2008
2018-01-22 - 2018-12-17: vol due to trade war
2018-12-18 - 2020-02-10: benefiting from Trade war w/ China: merging 2 and 3 together 
2020-02-11 - 2020-03-23: crash due to COVID
2020-03-24 - 2020-07-25: recovery due to Fed intervention

### Initiate lists

```python
startlist = ['2011-01-01','2018-01-22','2020-02-11','2020-03-24']
endlist = ['2018-01-21','2020-02-10','2020-03-23',date.today()]
description = ['Recovering from Great Recession (2011 to 2018)', 'Trade War period (2018 to 2020)',
            'COVID crash (02-2020 to 03-2020)', 'Recovery due to Fed intervention (03-2020 to 07-2020)']
tickerlist = ['SPY', 'IWM', 'DIA','TLT','QQQ','GLD']
recent_vol = []
```

### a. Define realised_vol function

```python
def realised_vol (tick, start_date, end_date, descr):
    ticker = web.DataReader(tick, data_source = 'yahoo', start = start_date, end = end_date)

#Calculate log returns
    ticker = ticker.reset_index()
    ticker_r = np.log(ticker['Close'])- np.log(ticker['Close'].shift(1))
    
#Calculate SMA
    ticker_rolling = ticker_r.rolling(window = 30)

#Calculate EWM 
    ticker_ewm = ticker_r.ewm(alpha = 0.06)

    ticker_MA = ticker_rolling.std()  #ma std
    ticker_EWMA = ticker_r.ewm(alpha=0.06).std()  #ewma std
    
    if end_date ==date.today():
        recent_vol.append(ticker_EWMA.iloc[-1])
    
#Draw realised vol
    fig, ax = plt.subplots(1,1,figsize = (10,6))
    ax.plot(ticker['Date'], ticker_MA, label='MA vol')
    ax.plot(ticker['Date'], ticker_EWMA,label='EWMA vol')
    ax.set_xlabel("Time"), ax.set_ylabel("Vol"), ax.set_title(tick+ " Realised Vol - "+ descr)
    plt.legend()
    fig.autofmt_xdate()
    plt.show()
```

### b. Call realised_vol function for all tickers, print out most realized vol and premiums

```python
for tick in tickerlist:
    for start_date, end_date, descr in zip(startlist, endlist, description): 
        realised_vol(tick, start_date, end_date, descr)
        
print('Most recent realised vol corresponding to', tickerlist,' are: ', recent_vol)

premium_yahoo = [dec_atm_yahoo[i] - recent_vol[i] for i in range(len(recent_vol))]
premium_newton = [dec_atm_newton[i] - recent_vol[i] for i in range(len(recent_vol))]

print('Premium using Yahoo realised vol corresponding to', tickerlist,' are: ',premium_yahoo)
print('Premium using Newton realised vol corresponding to', tickerlist,' are: ',premium_newton)

```

### c. Plot Vol premiums for each ticker

```python
figure = plt.figure()
plt.scatter(tickerlist, premium_yahoo, label='Yahoo vol premium')
plt.scatter(tickerlist, premium_newton, label='Newton vol premium')
plt.xlabel("Ticker"), plt.ylabel("Premium"), plt.title('Year end vol risk premium')
plt.legend()
```
